```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

def crawl_data(url):
    driver = webdriver.Chrome(executable_path='/Users/jeonghojin/chromedriver')
    driver.get(url)

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    data = {}

    for i in range(1, 25):  # Rows 1 and 2
        row_data = []  # Temporary list to store data for the current i value
        for col in range(1, 25):
            selector = f"#goingUpBody > tr:nth-child({i}) > td:nth-child({col})"
            element = soup.select_one(selector)
            
            selector1 = f"#goingUpName > tr:nth-child({i}) > td"
            name_element = soup.select_one(selector1)

            if not element or not name_element:  # If either element is not found, break the inner loop
                break

            key = name_element.text.strip()
            row_data.append(element.text.strip() if element else 'N/A')

        if not row_data:  # If row_data is empty, break the outer loop
            break

        # Add row data to the main data dictionary
        data[key] = row_data

    driver.quit()

    df = pd.DataFrame(data)
    return df

base_address = 'https://www.gjtic.go.kr/road-statistic'
result_df = crawl_data(base_address)

print(result_df)
```
